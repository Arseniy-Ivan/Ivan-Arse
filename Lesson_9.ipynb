```python
s = [("САРСик", 7, "Фрол", "Фролов"), ("Крепыш", 19, "Олег", "Ширяев"), ("Кексик", 6, "Никита", "Курочкин"), ("Пивасик", 3, "Фрол", "Фролов"), ("Захарчик", 4, "Никита", "Курочкин")]
dict = {}

for tuple in s:
    a,b,c,d = tuple
    if c+" "+d in dict:
        dict[c+" "+d]+= a,b
    else:
        dict[c+" "+d] = a,b
print(dict)



```

```python
import random

a = [random.randint(1, 1000) for _ in range(10)]

print(a)

for i in range (len(a)-1):
    for j in range (len(a)-1-i):
        if a[j] > a[j+1]:
            a[j], a[j+1] = a[j+1], a[j]

print(a)


```
