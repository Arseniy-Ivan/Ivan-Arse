```python
import random 
n = random.randint (1,100)
k=0

for i in range (10):
    p = float(input())
    k+=1
    if p == n:
        print("Умничка!")
        break
    else:
        k+=1
        print("Не умничка")
        if p > n:
            print("Меньше")
        else:
            print("Больше")
if k == 20:
    print("GAME OVER")
else:
    print ("ВЫИГРАЛ")
        
```
